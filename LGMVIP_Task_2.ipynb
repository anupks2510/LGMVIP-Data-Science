#Task 2 : Image to Pencil Sketch with Python

In this we have to convert a given image into its pencil sketch using python. In these we use CV ( Computer Vision ) library and Pillow module for the Image Processing.



#Importing Libraries

```python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

#Loading  Image

```python
img  =  cv2.imread("Image.jpg")
img = cv2.resize(img,(300,200))
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
```

#Converting Image to Greyscale Image¶

```python
greyscale_img =  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale_img))
```

#Converting the Image to Negative Image

```python
neg_img = cv2.bitwise_not(greyscale_img)
display(Image.fromarray(neg_img))
```

#Smoothening the Image

```python
smooth_img = cv2.GaussianBlur(neg_img, (5,5), 0, 0)
display(Image.fromarray(smooth_img))
```

#Converting Image into Pencil Sketch¶

```python
sketch_img = 255 - smooth_img
sketch = cv2.divide(greyscale_img, sketch_img, scale=250.0)
display(Image.fromarray(sketch))
```

#Comparision Between Original Image and Pencil Sketch

```python
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
display(Image.fromarray(sketch))
```
